```python
import matplotlib.pyplot as plt
import numpy as np

# Параметры векторов
I = 2  # Амплитуда тока (базовый вектор), совпадает с горизонтальной осью
R1 = 30  # Ом
Z_parallel = 30  # Ом
Z_L = 30j  # Ом
Z_C = -30j  # Ом

# Вычисление напряжений на элементах
U_R1 = I * R1
U_parallel = I * Z_parallel
U_R2 = U_parallel
U_L = I * Z_L
U_C = I * Z_C
U_total = U_R1 + U_parallel  # Общее напряжение

# Создание комплексных значений напряжений
vectors = {
    "I (Ток)": I + 0j,
    "U_R1": U_R1 + 0j,
    "U_parallel": U_parallel + 0j,
    "U_L": U_L,
    "U_C": U_C,
    "U_total": U_total + 0j,
}

# Параметры для построения
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Построение векторов
for label, vec in vectors.items():
    plt.arrow(0, 0, vec.real, vec.imag,
              head_width=5, head_length=5, fc='blue', ec='blue')
    plt.text(vec.real * 1.1, vec.imag * 1.1, f"{label} ({np.abs(vec):.1f} В)",
             fontsize=9, ha='center')

# Настройки графика
plt.xlim(-100, 150)
plt.ylim(-100, 100)
plt.xlabel("Re")
plt.ylabel("Im")
plt.title("Векторная диаграмма напряжений и токов")
plt.grid()

plt.show()
```
