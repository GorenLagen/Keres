```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Создаем диапазон значений x
x = np.linspace(0, 20, 5000)

# Вычисляем значения функции плотности вероятности (PDF) для хи-квадрат распределения
for df in [1, 2, 3, 4, 5]:
	pdf = chi2.pdf(x, df)

	# Строим график
	plt.plot(x, pdf, label=f'Хи-квадрат (df={df})')
	plt.title('Хи-квадрат распределение')
	plt.xlabel('x')
	plt.ylabel('Плотность вероятности')
	plt.legend()
	plt.grid(True)
	plt.show()
```

```python
df = 9
pdf = chi2.pdf(x, df)

# Строим график
plt.plot(x, pdf, label=f'Хи-квадрат (df={df})')
plt.title('Хи-квадрат распределение')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid(True)
plt.show()
```
