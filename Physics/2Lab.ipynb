```python
def calculate_variance(df, column_name):
    return 0.1 * np.abs(np.diff(df[column_name])) / 360

def process_data(df_mercury, df_neon):
    variance_mercury = calculate_variance(df_mercury, "–î–µ–ª–µ–Ω–∏—è –±–∞—Ä–∞–±–∞–Ω–∞")
    variance_neon = calculate_variance(df_neon, "–î–µ–ª–µ–Ω–∏—è –±–∞—Ä–∞–±–∞–Ω–∞")

    max_len = max(len(variance_mercury), len(variance_neon))
    result = pd.DataFrame(index=range(max_len))

    result["D_fi_mer"] = pd.Series(variance_mercury)
    result["D_l_mer"] = result["D_fi_mer"] * 4.9
    result["lambda mercury"] = df_mercury["–î–ª–∏–Ω–∞ –≤–æ–ª–Ω—ã (–Ω–º)"][:-1]

    result["D_fi_neon"] = pd.Series(variance_neon)
    result["D_l_neon"] = result["D_fi_neon"] * 4.9
    result["lambda neon"] = df_neon["–î–ª–∏–Ω–∞ –≤–æ–ª–Ω—ã (–Ω–º)"]

    result = result[["D_fi_mer", "D_l_mer", "D_fi_neon", "D_l_neon", "lambda mercury", "lambda neon"]]

    for col in result.columns:
        if result[col].dtype in ['float64', 'int64']:
            result[col] = result[col].round(3)

    return result

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
result = process_data(df_mercury, df_neon)
result = pd.DataFrame(result)
result
```
