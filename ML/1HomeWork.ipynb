# Упражнение 5

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

x = np.array([1, 2, 3, 4, 5])
y = np.array([2.1, 3.7, 5.6, 7.2, 8.9])
y_err = np.array([0.2, 0.3, 0.3, 0.4, 0.5])  

def linear(x, a, b):
    return a * x + b

def quadratic(x, a, b, c):
    return a * x**2 + b * x + c

popt_linear, _ = curve_fit(linear, x, y, sigma=y_err)
popt_quadratic, _ = curve_fit(quadratic, x, y, sigma=y_err)

x_fit = np.linspace(min(x), max(x), 100)
y_fit_linear = linear(x_fit, *popt_linear)
y_fit_quadratic = quadratic(x_fit, *popt_quadratic)

plt.figure(figsize=(10, 6))
plt.errorbar(x, y, yerr=y_err, fmt='o', label='Данные с погрешностями')
plt.plot(x_fit, y_fit_linear, 'r-', linewidth = 2, label='Линейная аппроксимация')
plt.plot(x_fit, y_fit_quadratic, 'g-', label='Квадратичная аппроксимация')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Аппроксимация данных')
plt.legend()
plt.grid(True)
plt.show()
```

# Упражнение 6

```python
a = 0.5
b = 3
n = 50  

def weierstrass(x, a, b, n):
    return sum(a**k * np.cos(b**k * np.pi * x) for k in range(n))

x = np.linspace(-2, 2, 1000)

y = [weierstrass(xi, a, b, n) for xi in x]

plt.plot(x, y, 'b-', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
